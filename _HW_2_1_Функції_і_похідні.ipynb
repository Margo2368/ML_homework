1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x = sp.symbols('x')

# Визначаємо функцію
f = 10 * x - 5
df = sp.diff(f, x)

print("Функція:", f)
print("Похідна:", df)
```

```python
x = np.linspace(0, 10, 200)
f = 10 * x - 5
df = 10 * x**0

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

plt.title("Графік функції оцінки та її похідної")
plt.xlabel("Години навчання")
plt.ylabel("Оцінка")
plt.legend()
plt.grid(True)

plt.show()
```

**Спостереження:**

Похідна від х = 10, це означає, що кожна година навчання збільшує оцінку на 10 балів.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



###𝑓(𝑥)=sin(𝑥)

```python
x = sp.symbols('x')

f_sin = sp.sin(x)
df_sin = sp.diff(f_sin, x)

print("Функція:", f_sin, " Похідна:", df_sin)
```

```python
xs = np.linspace(0, 10, 300)

plt.figure(figsize=(12, 5))
plt.plot(xs, np.sin(xs),  label='f(x) = sin(x)',  color='blue')
plt.plot(xs, np.cos(xs), '--', label="f'(x) = cos(x)", color='green')
plt.title("f(x) = sin(x) та її похідна")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

### 𝑓(𝑥)=ln(𝑥)

```python
x = sp.symbols('x')

f_ln = sp.ln(x)
df_ln = sp.diff(f_ln, x)

print("Функція:", f_ln, " Похідна:", df_ln)
```

```python
xs_ln = np.linspace(0.1, 10, 300)

plt.figure(figsize=(12, 5))
plt.plot(xs_ln, np.log(xs_ln), label='f(x) = ln(x)', color='blue')
plt.plot(xs_ln, 1 / xs_ln, '--', label="f'(x) = 1/x", color='green')
plt.title("f(x) = ln(x) та її похідна")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

f = x**2 - 5*x + 1
df = sp.diff(f, x)

print("Функція витрат f(x) =", f)
print("Похідна f'(x) =", df)
```

```python
xs = np.linspace(0, 100, 300)

f_vals  = xs**2 - 5*xs + 1
df_vals = 2*xs - 5

plt.figure(figsize=(12, 5))
plt.plot(xs, f_vals, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(xs, df_vals, '--', label="f'(x) = 2x - 5", color='green')
plt.title("Функція загальних витрат")
plt.xlabel("Кількість вироблених одиниць")
plt.ylabel("Витрати")
plt.legend()
plt.grid(True)

plt.show()
```
